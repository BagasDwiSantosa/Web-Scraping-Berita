```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
search = str(input("Cari : "))
```

```python
url = "https://www.liputan6.com/search?q="+search
```

```python
r = requests.get(url)
soup = BeautifulSoup(r.text, 'html.parser')
soup
```

```python
data_berita = pd.DataFrame()
```

```python
for p in soup.find_all("article", {"class":"articles--iridescent-list--item articles--iridescent-list--text-item"}):
    #Judul
    if str(p.find("h4")) != "None":
        title = p.find("h4").get_text().lower()
        # Cleansing
        title = title.replace("\n", "")
    # kategori
    if str(p.find("aside", {"class":"articles--iridescent-list--text-item__details"})) != "None":
        kategori = p.find("a", {"class":"articles--iridescent-list--text-item__category"}).get_text().lower()
    # Tanggal
    if str(p.find("span", {"class":"articles--iridescent-list--text-item__datetime"})) != "None":
        dates = p.find("time", {"class": "articles--iridescent-list--text-item__time timeago"}).get_text()
        dates = dates.split(',')
        date = dates[0]

    data_berita = data_berita.append({'Judul': title, 'Kategori': kategori}, ignore_index=True)
```
