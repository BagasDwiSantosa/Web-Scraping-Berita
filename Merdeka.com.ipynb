```python
from bs4 import BeautifulSoup
import requests

page = str(input("Page : "))
# url = "https://www.merdeka.com/trending/index2"
url = "https://www.merdeka.com/trending/index"+page
response = requests.get(url)

response
```

```python
import csv
import requests
from bs4 import BeautifulSoup
import pandas as pd
from multiprocessing import Pool
```

```python
soup = BeautifulSoup ( response.content, "html.parser")
articles = soup.find('div', class_='inner-content')
p = articles.find_all('div', class_='meta-content')
p
```

```python
titles = []
img = []
link =[]

for row in p :
    title = row.find('h3').find( 'a')
    title_element = title.text.strip()
    titles.append(title_element)

    articleLink_element = row.find('a')
    articleLink = articleLink_element['href']
    link.append(articleLink)

    linkImg_element = row.find('img').attrs['src']
    img.append(linkImg_element)
```

```python
data = {'Judul': titles,
        'Link Img' : img,
        'Link': link}
import pandas as pd
df = pd.DataFrame(data)

df.head()
```
