```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin
```

```python
def scrape_website(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    articles = soup.find('div', class_="latest__wrap")
    articles = articles.find_all('div', class_="latest__item")
    articles

    scraped_data = []
    for article in articles:
      title_element = article.find('h2').find( 'a')
      title = title_element.text.strip()
      # scraped.append(title)

      link_element = article.find('h2', 'latest__title')
      link = link_element.find('a', 'latest__link').attrs['href']
      # link = link_element.text.strip()

      img_element_1=article.find('div',class_="latest__img")
      img_element_2=(img_element_1.find('a'))
      img = img_element_2.find("img").attrs['data-src']
      # images.append(img)

      date_element_1 = article.find('div','latest__right')
      date_element_2 = date_element_1.find('date', 'latest__date')
      date = date_element_2.text.strip()
      # dates.append(date)

      scraped_data.append({
            'Image': img,
            'Link' : link,
            'Judul': title,
            'Tanggal': date
        })
    return scraped_data
```

```python
url = 'https://food.indozone.id/'
```

```python
scraped_data = scrape_website(url)
```
