Import Module

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

Search Query

```python
search = input('Search Berita : ')
```

```python
url = 'https://www.cnbcindonesia.com/search?query='+search
```

Get HTML

```python
req = requests.get(url)
soup = BeautifulSoup(req.text, 'html.parser')
```

```python
# html = soup.find('div' , class_="lm_content mt10")
html = soup.find('ul', class_='list media_rows middle thumb terbaru gtm_indeks_feed')
articles = html.find_all('article')
articles
```

```python
scrape = []

for article in articles:
  title_element = article.find('div', class_='box_text').find('h2')
  title = title_element.get_text()

  link = article.find('a').attrs['href']

  date_element = article.find('div', class_='box_text').find('span', class_='date')
  date = date_element.get_text()

  imglink_element = article.find('span',  class_='ratiobox_content lqd')
  imglink = imglink_element.find('img').attrs['src']

  scrape.append({
      'Title':title,
      'Link':link,
      'Date':date,
      'imglink':imglink
  })
```

```python
df = pd.DataFrame(scrape)
df
```
