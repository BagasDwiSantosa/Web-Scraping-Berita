```python
from bs4 import BeautifulSoup
import requests

url = "https://www.detik.com/search/searchall?query=terkini&siteid=2"
response = requests.get(url)

response
```

```python
soup = BeautifulSoup ( response.text, "html.parser")
html = soup.find('div' , {"class" :"list media_rows list-berita"})
p = html.find_all('article')
```

```python
dates = []
titles= []
ket=[]

for row in p :
    date = row.find("span", {"class" : "date"})
    dates.append(date.text)

    title = row.find("h2", {"class" : "title"})
    print(title)
    titles.append(title.text)
    pp = row.find("p")
    ket.append(pp.text)
```

```python
data = {'satu': dates,
        'dua' : titles,
        'tiga': ket}
import pandas as pd
df = pd.DataFrame(data)

df.head()
```

```python
import pandas as pd

df = pd.DataFrame ({'Katogri' : dates, 'Judul' :  titles})
df.to_csv('/content/drive/MyDrive/DataSet/HasilScraping/data.csv', index = True, encoding = 'utf-8')
```
